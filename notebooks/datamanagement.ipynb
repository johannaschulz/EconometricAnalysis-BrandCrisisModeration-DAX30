### Aggregation of the data, monthly frequency | Yahoo Finance
In the next step, the data is aggregated to a monthly frequency. This is done to reduce the noise in the data and to be able to compare the stock prices of the different companies. The aggregation is done by taking the Volume-Weighted-Average-Price (VWAP) of the stock prices of each month.

```python
import pandas as pd

dataset['Date'] = pd.to_datetime(dataset['Date'])

# 'MonthYear' column because I need to be sure that each month is assigned to a year
dataset['MonthYear'] = dataset['Date'].dt.to_period('M')

# Calculate Price*Volume and then group by 'Company' and 'MonthYea', I will use the closing price (reasons see master thesis docs)
dataset['Price_Volume'] = dataset['Close'] * dataset['Volume']
grouped = dataset.groupby(['Company', 'MonthYear'])

# Calculate VWAP for each group (further information see Chapter 4.1 Operationalisation))
monthly_vwap_dataset = grouped.apply(lambda x: x['Price_Volume'].sum() / x['Volume'].sum() if x['Volume'].sum() > 0 else 0)

# Reset index to make 'Company' and 'Month-Year' columns
monthly_vwap_dataset = monthly_vwap_dataset.reset_index()
monthly_vwap_dataset.rename(columns={0: 'VWAP'}, inplace=True)

## The resulting 'monthly_vwap' DataFrame will have the VWAP values for each company for each month

# print the head of the dataset
print(monthly_vwap_dataset.head())

# save dataset as csv 
monthly_vwap_dataset.to_csv('../outputs/monthly_vwap_dataset.csv', index=False)
```

## Data set: CPI Germany

As an extend to the analysis, additonal data about the Consumer Price Index is added. This will give insights into the effect of inflation on the stock prices. CPI reflects changes in the price level of a market basket of consumer goods and services purchased by households. Theoretically, inflation (measured by CPI) can affect stock prices (VWAP) through various channels like consumer purchasing power, company costs, and overall economic health.

The data was downloaded from the Federal Statistical Office of Germany (Statistisches Bundesamt). The data is available on a monthly basis from 1991 until date of data import. The data is imported from the csv file. Only the time period of the analysis (January 2019 until December 2020), will be used. 

It is important to note that the values of the Column CPI is based on the December 2019 = 100. This means that the values are relative to the the month December 2019. 

```python
print(os.listdir('../data/CPI/'))
```

```python
CPI_data = 'CPI_data.xlsx'  

excel_file_path = f'../data/CPI/{CPI_data}'

CPI_data = pd.read_excel(excel_file_path, header=5)

CPI_data.head()
```

```python
# change , to . in CPI, CPI_Change_YoY and CPI_Change_MoM
CPI_data['CPI'] = CPI_data['CPI'].str.replace(',', '.')
CPI_data['CPI_Change_YoY'] = CPI_data['CPI_Change_YoY'].str.replace(',', '.')
CPI_data['CPI_Change_MoM'] = CPI_data['CPI_Change_MoM'].str.replace(',', '.')
CPI_data.head()
```

### Final Variable Creation of the final data set 

#### Definition of "crisis" and "no crisis" period

```python
# 'Date' is a datetime
monthly_vwap_brand_data['MonthYear'] = pd.to_datetime(monthly_vwap_brand_data['MonthYear'])

# Defintion of start and end of the crisis period
crisis_start = pd.to_datetime('2020-01-01')
crisis_end = pd.to_datetime('2020-12-31')

# Creation  'Crisis' dummy variable
monthly_vwap_brand_data['Crisis'] = monthly_vwap_brand_data['MonthYear'].apply(lambda x: 1 if crisis_start <= x <= crisis_end else 0)

monthly_vwap_brand_data.head()

# save monthly_vwap_brand_data as xlsx
monthly_vwap_brand_data.to_excel('../outputs/monthly_vwap_brand_data.xlsx', index=False)
```

Furthermore, a lagged VWAP variable is created to use as a control variable to account for the autocorrelation of the stock prices. The lagged variable is created by shifting the VWAP variable by one month.

```python
monthly_vwap_brand_data = monthly_vwap_brand_data.sort_values(by=['CompanyName', 'MonthYear'])

# Group by 'CompanyName' and create the lagged VWAP variable
monthly_vwap_brand_data['VWAP_lagged'] = monthly_vwap_brand_data.groupby('CompanyName')['VWAP'].shift(1)

monthly_vwap_brand_data.head()
```

### Control Variables

The information on sector and industry will be used as control variables in the econometric analysis. The sector and industry information is manually taken from the Yahoo Finance data set. For the 30 companies, there exist 9 different sectors and 23 different industries. However, given the use of a fixed effects model in the panel data analysis, it was decided to exclude detailed industry information from the primary model (without fixed effects). This decision is based on the fact that both the industry and sector classifications are time-invariant characteristics that are considered and made redundant in the later fixed effects analysis. As a result, the analysis will continue with the available sector data in its current form. 

I export sector and industry information to include it in my thesis appendix to give evidence for my decision to exclude industry information from the primary model and to report my decision. 

```python
# Removing duplicate companies for accurate sector and industry counts
industry_sector_data = monthly_vwap_brand_data.drop_duplicates(subset=['CompanyName'])

# Counting the number of companies in each sector
sector_counts = industry_sector_data.groupby('Sector').size().reset_index(name='Number of Companies')

# Counting the number of companies in each industry
industry_counts = industry_sector_data.groupby('Industry').size().reset_index(name='Number of Companies')

print("Sector Counts:")
print(sector_counts)

print("\nIndustry Counts:")
print(industry_counts)

# save sector_counts as xlsx
sector_counts.to_excel('../outputs/sector_counts.xlsx', index=False)

# save industry_counts as xlsx
industry_counts.to_excel('../outputs/industry_counts.xlsx', index=False)
```
